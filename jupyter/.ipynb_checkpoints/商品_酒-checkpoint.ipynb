## 练习

###  数据分组-探索酒类消费数据

#### 1.将数据框命名为drinks 

```python
import pandas as pd
import numpy as np
```

```python
drinks = pd.read_csv("../data/drinks.csv")
drinks.head(30)
```

#### 2.哪个大陆平均消耗的啤酒最多

```python
drinks.groupby("continent").agg({"beer_servings":np.mean}).sort_values("beer_servings")
```

#### 3.打印出每个大陆红酒消耗的描述性统计

```python
drinks.groupby("continent")["wine_servings"].describe()
```

#### 4.打印出每个大陆每种酒的消耗平均值

```python
drinks[["beer_servings","spirit_servings","wine_servings","continent"]].groupby("continent").agg(np.mean)
```

#### 5.打印出每个大陆每种酒的消耗中位数

```python
drinks[["beer_servings","spirit_servings","wine_servings","continent"]].groupby("continent").agg(np.median)
```

#### 6.打印出每个大陆对spirit饮品消耗的平均值最大值最小值

```python
drinks[["spirit_servings","continent"]].groupby("continent").agg([np.mean,np.max,np.min])
```
