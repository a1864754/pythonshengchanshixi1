### 柱形图

```python
from pyecharts.charts import Bar
from pyecharts import options as opts# 导入主题设置
from pyecharts.globals import ThemeType # 导入主题设置

bar = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))
    .add_xaxis(['衬衫','羊毛衫','雪纺衫','裤子','高跟鞋','袜子'])
    .add_yaxis('商家A',[5,20,30,10,75,90])
    .add_yaxis('商家B',[15,6,36,40,30,60])
    .add_yaxis('商家C',[12,62,25,45,31,40])
    .set_global_opts(title_opts = opts.TitleOpts(title='这是一个柱形图',subtitle='这是一个副标题'))
      )
bar.render_notebook()
bar.render('./pyechart实例图.html') # 可以保存为网页
```

### 折线图 

```python
from pyecharts.charts import Line
from pyecharts import options as opt
from pyecharts.globals import ThemeType
from pyecharts.faker import Faker
bar = (
    Line(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))
    .add_xaxis(['衬衫','羊毛衫','雪纺衫','裤子','高跟鞋','袜子'])
    .add_yaxis('商家A',Faker.values(),is_smooth=True) #is_smooth=True 开启平滑—绘制曲线图
    .add_yaxis('商家B',Faker.values(),is_smooth=True)
    .set_global_opts(title_opts = opts.TitleOpts(title='这是一个折线图',subtitle='这是一个副标题'),
    tooltip_opts = opts.TooltipOpts(trigger='axis'))# 提示框组件
      )
bar.render_notebook()
```

#### 极坐标

```python
import math

from pyecharts import options as opts
from pyecharts.charts import Polar



import numpy as np
import matplotlib.pyplot as plt
 
theta = np.linspace(0, 2*np.pi, 1000)
y = np.pi * (1 - np.sin(theta))


data = []
for i in theta:
    theta =i
    r = np.pi * (1 - np.sin(i))
    data.append([r, theta])
# hour = [i for i in range(1, 250)]
print(data)
c = (
    Polar()
    .add("love", data)
    .set_global_opts(title_opts=opts.TitleOpts(title="Polar-Love"))
    
)
c.render_notebook()
```

#### 多子图饼图 

```python
from pyecharts.charts import Pie
source=[
            ["product", "2012", "2013", "2014", "2015", "2016", "2017"],
            ["Matcha Latte", 41.1, 30.4, 65.1, 53.3, 83.8, 98.7],
            ["Milk Tea", 86.5, 92.1, 85.7, 83.1, 73.4, 55.1],
            ["Cheese Cocoa", 24.1, 67.2, 79.5, 86.4, 65.2, 82.5],
            ["Walnut Brownie", 55.2, 67.1, 69.2, 72.4, 53.9, 39.1],
        ]
```

```python
c = (Pie(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))
     .add_dataset(source)
     .add(
         series_name = 'Matcha Latte',
         data_pair = [],
         radius = 60,
         center = ["25%","25%"],
         encode = {"itemName":"product","value":"2012"}            
      )
     .add(
         series_name = 'Milk Tea',
         data_pair = [],
         radius = 60,
         center = ["25%","75%"],
         encode = {"itemName":"product","value":"2013"}            
      )
     .add(
         series_name = 'Cheese Cocoa',
         data_pair = [],
         radius = 60,
         center = ["75%","25%"],
         encode = {"itemName":"product","value":"2014"}            
      )
     .add(
         series_name = 'Walnut Brownie',
         data_pair = [],
         radius = 60,
         center = ["75%","75%"],
         encode = {"itemName":"product","value":"2016"}            
      )
     .set_global_opts(title_opts=opts.TitleOpts(title='这是一堆饼图'))
     )
c.render_notebook()
```

#### Geo绘图 

```python
from pyecharts.charts import Geo
from pyecharts.globals import SymbolType # 点标记
from pyecharts.globals import ThemeType,ChartType # 主题
```

```python
c = (Geo()
    .add_schema(maptype="china")
    .add("",[("广州",55),("西藏",20),("北京",80),("陕西",40),("重庆",100)],
     type_=ChartType.EFFECT_SCATTER,
     color = 'write',
        )
     .add("geo",
          [("广州","西藏"),("广东","西安"),("广东","重庆"),("广东","黑龙江"),("广东","内蒙古")],
          type_ = ChartType.LINES,
          effect_opts = opts.EffectOpts(
              symbol = SymbolType.ARROW,
              symbol_size=5,
              color="blue"
          )    
     )
        
    )
c.render_notebook()
```

```python
colors = ["#5793f3", "#d14a61", "#675bba"]
x_data = ["1月", "2月", "3月", "4月", "5月", "6月", "7月", "8月", "9月", "10月", "11月", "12月"]
legend_list = ["蒸发量", "降水量", "平均温度"]
evaporation_capacity = [2.0,4.9,7.0,23.2,25.6,76.7,135.6,162.2,32.6,20.0,6.4,3.3,]
rainfall_capacity = [2.6,5.9,9.0,26.4,28.7,70.7,175.6,182.2,48.7,18.8,6.0,2.3,]
average_temperature = [2.0, 2.2, 3.3, 4.5, 6.3, 10.2, 20.3, 23.4, 23.0, 16.5, 12.0, 6.2]
```

```python
# 生成条形图
bar =(Bar(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))
     .add_xaxis(x_data) # x轴数据
     .add_yaxis(series_name="蒸发量",
               yaxis_data=evaporation_capacity,
               label_opts = opts.LabelOpts(is_show=False))
     .add_yaxis(series_name="降雨量",
               yaxis_data=rainfall_capacity,
               label_opts = opts.LabelOpts(is_show=False))
      # 生成右轴
     .extend_axis(
               yaxis = opts.AxisOpts(name="温度",
                               type_="value",
                               min_=0,
                               max_=25,
                               interval=5,
                               axislabel_opts=opts.LabelOpts(formatter="{value} ℃"))
     )
     .set_global_opts(
         title_opts=opts.TooltipOpts(
             is_show=True,trigger='axis',axis_pointer_type='cross'),
         
         xaxis_opts = opts.AxisOpts(
             type_ = 'category',
             axispointer_opts = opts.AxisPointerOpts(is_show=True,type_='shadow')),
         
         # 设置左轴
         yaxis_opts = opts.AxisOpts(
             name='水量',
             type_='value',
             min_=0,
             max_=250,
             interval=50,
             axislabel_opts=opts.LabelOpts(formatter="{value} ml"),
             axistick_opts = opts.AxisTickOpts(is_show=True),
             splitline_opts = opts.SplitLineOpts(is_show=True)
             )
         )

     )
# 生成折线
line = (Line()
       .add_xaxis(x_data)
       .add_yaxis(series_name="平均温度",
                 yaxis_index=1,
                 y_axis=average_temperature,
                 label_opts = opts.LabelOpts(is_show=False))
       )
bar.overlap(line).render_notebook()
```
